```python
import numpy as np
import statistics as stats

list_of_numbers = [10, 20, 30, 40, 50]

mean_value = np.mean(list_of_numbers)
print("Mean value:", mean_value)

median_value = np.median(list_of_numbers)
print("Median value:", median_value)   

std_deviation = np.std(list_of_numbers)
print("Standard Deviation:", std_deviation)

variance_value = np.var(list_of_numbers)
print("Variance:", variance_value)

mod_value = stats.mode(list_of_numbers)
print("Mode value:", mod_value)

q1 = np.percentile(list_of_numbers, 25)
print("First Quartile (Q1):", q1)

q2 = np.percentile(list_of_numbers, 50)
print("Second Quartile (Q2):", q2)

q3 = np.percentile(list_of_numbers, 75)
print("Third Quartile (Q3):", q3)

iqr = q3 - q1
print("Interquartile Range (IQR):", iqr)

```
